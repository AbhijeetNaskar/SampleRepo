```python
x1 = [3, 5 , 9]
y1 = [22, 33, 54]
x2 = [4, 7, 8]
y2 = [11, 35, 76]

#adding style to the plot
plt.style.use("fivethirtyeight")

#showing multiple graph in the same plot
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
x1 = [3, 5 , 9]
y1 = [22, 33, 54]
x2 = [4, 7, 8]
y2 = [11, 35, 76]

#adding style to the plot
plt.style.use("ggplot")

#showing multiple graph in the same plot, using label legend
plt.plot(x1, y1, label="first")
plt.plot(x2, y2, label="second")
plt.legend()
plt.show()
```
